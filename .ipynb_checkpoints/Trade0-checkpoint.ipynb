```python
class Environment:
    DF_PATH = 'GOLD.csv'
    INITIAL_MONEY = 100000
    INITIAL_STOCKS = 0
    #OBSERVATION_SPACE = 6
    ACTION_SPACE = 3
    WINDOW_SIZE = 32
    COMISSION = 100
    
    
    def __init__(self, train_test_split=0.8):
        self.df = pd.read_csv(self.DF_PATH)
        self.df = self.df.dropna()
        self.df.pop('<DATE>')
        self.df.pop('<TIME>')
        #if 'Date' in self.df:
        #    self.df.pop('Date')
        self.train_test_split=train_test_split
        self.data = np.array(self.df[:int(len(self.df.values)*train_test_split)])
        #self.test_data = self.df.Close[int(len(self.df.values)*train_test_split):]
        self.current_money = self.INITIAL_MONEY
        self.current_stocks = self.INITIAL_STOCKS
        self.cursor = self.WINDOW_SIZE - 1
        
    def reset(self, test=False):
        if test:
            self.data = np.array(self.df[int(len(self.df.values)*self.train_test_split):])
        else:
            self.data = np.array(self.df[:int(len(self.df.values)*self.train_test_split)])
        self.current_money = self.INITIAL_MONEY
        self.current_stocks = self.INITIAL_STOCKS
        self.cursor = self.WINDOW_SIZE - 1
        
    def sell(self):
        price = self.data[self.cursor][3]
        sold = self.current_stocks * price
        self.current_stocks = 0
        self.current_money += sold - self.COMISSION
        return self.current_money - self.INITIAL_MONEY
    
    def hold(self):
        return 0
    
    def buy(self):
        price = self.data[self.cursor][3]
        n_stocks = int(self.current_money/price)
        if self.current_money < 0:
            n_stocks = 0
        self.current_money -= price*n_stocks + self.COMISSION
        self.current_stocks = n_stocks
        return 0
            
    def reward(self, profit):
        price = self.data[self.cursor][3]
        return self.current_money + price*self.current_stocks + profit
    
    def get_state(self):
        state = self.data[self.cursor-self.WINDOW_SIZE+1:self.cursor+1]
        return state
        
    def step(self, action):
        action -= 1
        
        while np.isnan(self.get_state()).any():
            self.cursor += 1
            print("JUST ESCAPED NAN")
        
        if action == -1:
            profit = self.sell()
        elif action == 0:
            profit = self.hold()
        elif action == 1:
            profit = self.buy()
            
        self.cursor += 1
            
        done = self.cursor == len(self.data) - 1
            
        return self.get_state(), self.reward(profit), done
```

```python
import sys
import pylab
import numpy as np
from tensorflow.keras.layers import Dense, CuDNNLSTM, Conv1D, Conv2D, MaxPool2D, Flatten, MaxPool1D, Reshape
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam

EPISODES = 20
TEST = False
TRAIN_TEST_SPLIT = 0.02


# A2C(Advantage Actor-Critic) agent for the Cartpole
class A2CAgent:
    def __init__(self, state_size, action_size):
        
        # get size of state and action
        self.state_size = state_size
        self.action_size = action_size
        self.value_size = 1

        # These are hyper parameters for the Policy Gradient
        self.discount_factor = 0.99
        self.actor_lr = 0.0001
        self.critic_lr = 0.0005

        # create model for policy network
        self.actor = self.build_actor()
        self.critic = self.build_critic()

        #if self.load_model:
        #    self.actor.load_weights("./save_model/cartpole_actor.h5")
        #    self.critic.load_weights("./save_model/cartpole_critic.h5")

    # approximate policy and value using Neural Network
    # actor: state is input and probability of each action is output of model
    def build_actor(self):
        actor = Sequential()
        #actor.add(Input(shape=(self.state_size,)))
        actor.add(Conv2D(8, kernel_size=(1, 5), input_shape=(env.WINDOW_SIZE, 5, 1), activation='relu',
                        kernel_initializer='he_uniform', padding='valid'))
        actor.add(Reshape((env.WINDOW_SIZE, 8)))
        actor.add(Conv1D(32, kernel_size=8, activation='relu',
                        kernel_initializer='he_uniform', padding='same'))
        actor.add(Conv1D(32, kernel_size=env.WINDOW_SIZE//2, activation='relu',
                        kernel_initializer='he_uniform', padding='valid'))
        actor.add(CuDNNLSTM(32, kernel_initializer='he_uniform'))
        actor.add(Flatten())
        actor.add(Dense(16, activation='relu',
                        kernel_initializer='he_uniform'))
        actor.add(Dense(16, activation='relu',
                        kernel_initializer='he_uniform'))
        actor.add(Dense(self.action_size, activation='softmax',
                        kernel_initializer='he_uniform'))
        actor.summary()
        # See note regarding crossentropy in cartpole_reinforce.py
        actor.compile(loss='categorical_crossentropy',
                      optimizer=Adam(lr=self.actor_lr))
        return actor

    # critic: state is input and value of state is output of model
    def build_critic(self):
        actor = Sequential()
        #actor.add(Input(shape=(self.state_size,)))
        actor.add(Conv2D(8, kernel_size=(1, 5), input_shape=(env.WINDOW_SIZE, 5, 1), activation='relu',
                        kernel_initializer='he_uniform', padding='valid'))
        actor.add(Reshape((env.WINDOW_SIZE, 8)))
        actor.add(Conv1D(32, kernel_size=8, activation='relu',
                        kernel_initializer='he_uniform', padding='same'))
        actor.add(Conv1D(32, kernel_size=env.WINDOW_SIZE//2, activation='relu',
                        kernel_initializer='he_uniform', padding='valid'))
        actor.add(CuDNNLSTM(32, kernel_initializer='he_uniform'))
        actor.add(Flatten())
        actor.add(Dense(16, activation='relu',
                        kernel_initializer='he_uniform'))
        actor.add(Dense(16, activation='relu',
                        kernel_initializer='he_uniform'))
        actor.add(Dense(1, activation='softmax',
                        kernel_initializer='he_uniform'))
        actor.summary()
        actor.compile(loss="mse", optimizer=Adam(lr=self.critic_lr))
        return actor

    # using the output of policy network, pick action stochastically
    def get_action(self, state):
        policy = self.actor.predict(state, batch_size=1).flatten()
        if np.isnan(policy).any():
            policy = [[1/3, 1/3, 1/3]]
            print("!!!somenan", policy, state)
        return np.random.choice(self.action_size, 1, p=policy)[0]

    # update policy network every episode
    def train_model(self, state, action, reward, next_state, done):
        target = np.zeros((1, self.value_size))
        advantages = np.zeros((1, self.action_size))
        

        value = self.critic.predict(np.reshape(state, [1, env.WINDOW_SIZE, 5, 1]))[0][0]
        next_value = self.critic.predict(np.reshape(next_state, [1, env.WINDOW_SIZE, 5, 1]))[0][0]
        
        #print(reward, self.discount_factor, next_value, value)

        if done:
            advantages[0][action] = reward - value
            target[0][0] = reward
        else:
            advantages[0][action] = reward + self.discount_factor * (next_value) - value
            target[0][0] = reward + self.discount_factor * next_value

        self.actor.fit(state, advantages, epochs=1, verbose=0)
        self.critic.fit(state, target, epochs=1, verbose=0)


if __name__ == "__main__":
    # In case of CartPole-v1, maximum length of episode is 500
    env = Environment(train_test_split=TRAIN_TEST_SPLIT)
    #for i in range(len(env.df)): env.df.loc[i, 'Close'] = float(env.df.loc[i, 'Close'].replace('.0000000', ''))
    #env.df.Close = env.df.Close.astype(float)
    # get size of state and action from environment
    state_size = env.WINDOW_SIZE
    action_size = env.ACTION_SPACE

    # make A2C agent
    agent = A2CAgent(state_size, action_size)

    scores, episodes = [], []
    agent.actor.trainable = not TEST
    agent.critic.trainable = not TEST

    for e in range(EPISODES):
        done = False
        score = 0
        env.reset(test=TEST)
        state = env.get_state()
        state = preprocessing.minmax_scale(state)
        state = np.reshape(state, [1, env.WINDOW_SIZE, 5, 1])
        c = 0
        stats = [0, 0, 0]
        buys = []
        sells = []
        while not done:
            
            c += 1
            print('\r%s' % c, end='')
            action = agent.get_action(state)
            if action == 0:
                if env.current_stocks > 0:
                    sells.append(env.cursor)
            if action == 2:
                if env.current_money > env.data[env.cursor][3]:
                    buys.append(env.cursor)
            stats[action] += 1
            next_state, reward, done = env.step(action)
            next_state = preprocessing.minmax_scale(next_state)
            next_state = np.reshape(next_state, [1, env.WINDOW_SIZE, 5, 1])
            #print(env.current_money)
            #print(len(env.current_stocks))
            agent.train_model(state, action, reward, next_state, done)

            score += reward
            state = next_state

            if done:
                # every episode, plot the play time
                #score = score if score == 500.0 else score + 100
                scores.append(env.current_money + env.current_stocks*env.data[env.cursor][3])
                episodes.append(e)
                pylab.plot(episodes, scores, 'b')
                pylab.savefig("cartpole_a2c.png")
                pylab.close()
                print("episode:", e, "  score:", env.current_money + env.current_stocks*env.data[env.cursor][3])
                pylab.bar(range(len(stats)), stats)
                pylab.savefig("stats.png")
                pylab.close()
                pylab.plot(env.data[:,3])
                pylab.plot(env.data[:,3], '^', alpha=0.4, markersize=10, color='m', label = 'buying signal', markevery = buys)
                pylab.plot(env.data[:,3], 'v', alpha=0.4, markersize=10, color='k', label = 'selling signal', markevery = sells)
                pylab.savefig("behavior.png")
                pylab.close()
                # if the mean of scores of last 10 episode is bigger than 490
                # stop training
                #if np.mean(scores[-min(10, len(scores)):]) > 10000:
                #    break

        # save the model
        if (e+1) % 5 == 0:
            agent.actor.save("a2ca.h5")
            agent.critic.save("a2cc.h5")
```
